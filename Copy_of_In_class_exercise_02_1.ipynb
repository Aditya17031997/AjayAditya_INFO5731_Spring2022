```python
import requests
from bs4 import BeautifulSoup #imported BeautifulSoap, Requests and pandas packages
import pandas as pd



url = 'https://www.imdb.com/search/title/?title_type=feature' #Fetched the data from this url
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml') #using BeautifulSoup package parsed the data from html
m_tags = soup.find_all('a', attrs={'class': None})
m_tags_data = [tag.attrs['href'] for tag in m_tags 
              if tag.attrs['href'].startswith('/title') & tag.attrs['href'].endswith('/')]
m_tags_data = list(dict.fromkeys(m_tags_data)) # convert to list
m_links = ["https://www.imdb.com" + tag + 'reviews' for tag in m_tags_data] # concatinating the url with the tags
print(m_links)

def minMax(a):
    print(len(a))
    minpos = a.index(min(a))
    maxpos = a.index(max(a))
    return minpos,maxpos

reviews_list = []
Rating = []
Rating1=[]
user_id=[]
movie_name=[]
short_descrip=[]
#for i in range(0,len(m_links)):
print(m_links[0])
print(m_links[1])
for i in range(0,50):
    soup = BeautifulSoup(requests.get(m_links[i]).text, 'html.parser')
    #print(soup)


    review_ratings = [tag.previous_element for tag in 
                           soup.find_all('span', attrs={'class': 'point-scale'})]

    if len(review_ratings) > 0:
        n_index, p_index = minMax(list(map(int, review_ratings)))
        review_list = soup.find_all('a', attrs={'class':'title'})
        reviews = ["https://www.imdb.com" + review['href'] for review in review_list]
    else:
        print(None)




    for j in reviews:
        text=BeautifulSoup(requests.get(j).text, 'html.parser').find('div', attrs={'class': 'text show-more__control'}).get_text()
        rat=BeautifulSoup(requests.get(j).text, 'html.parser').find('div', attrs={'class': 'lister-item-content'}).get_text()
        movie_user=(BeautifulSoup(requests.get(j).text, 'html.parser').find('title').get_text())
        short_des=(BeautifulSoup(requests.get(j).text, 'html.parser').find('a', attrs={'class': 'title'}).get_text())
        len_rat1=len(Rating1)
        #print(text.replace(' ','').replace('\n',''))
        #print(rat.replace(' ',''))
        if(text == ''):
            reviews_list.append('No Review')
        else:
            reviews_list.append(text)
        Rating.append(rat)

        user_id.append(movie_user.split("'s")[0])
        movie_name.append(movie_user.split('Review of')[1])
        short_descrip.append(short_des)
  
    print('Rating : '+str(len(Rating))+' review list:'+str(len(reviews_list)))
    if(len_rat1==0):
        k=0
    else:
        k=len_rat1
    for j in range(k,len(Rating)):
        Rating1.append(Rating[j].split('/10')[0].split('\n')[len(Rating[j].split('/10')[0].split('\n'))-1])
    print('rating1:'+str(len(Rating1)))

#print(len(reviews_list))
#print(len(Rating1))
data = {
        "Movie name" : movie_name,
        "User Name" : user_id,
        "Review" : short_descrip,
        'Detail Reviews': reviews_list,
        'Rating': Rating1,
        
        }
df = pd.DataFrame(data)
df

```
